```python
# 1. Build the linear model, fit it on data and print the summary of the errors
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import scale
import sklearn.linear_model as skl_lm
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
import statsmodels.formula.api as smf
import scipy.stats as spst

import seaborn as sns

credit = pd.read_csv('credit.csv')
#print(credit)
dummycredit=pd.read_csv('credit.csv')

for i in np.arange(400):
    if dummycredit['Gender'][i]=="Male": dummycredit['Gender'][i]=1 
    else: dummycredit['Gender'][i]=0
for i in np.arange(400):
    if dummycredit['Student'][i]=="Yes": dummycredit['Student'][i]=1 
    else: dummycredit['Student'][i]=0
for i in np.arange(400):
    if dummycredit['Married'][i]=="Yes": dummycredit['Married'][i]=1 
    else: dummycredit['Married'][i]=0
for i in np.arange(400):
    if dummycredit['Ethnicity'][i]=="Caucasian": dummycredit['Ethnicity'][i]=2 
    if dummycredit['Ethnicity'][i]=="African American": dummycredit['Ethnicity'][i]=1  
    if dummycredit['Ethnicity'][i]=="Asian": dummycredit['Ethnicity'][i]=0    
        
lm5 = smf.ols('Balance ~  Income+Limit+Rating+Cards+Age+Education+Gender+Student+Married+Ethnicity', data=dummycredit).fit()
print(lm5.summary())
```
